```python
data['number of family members'] = data['number of children'] + data['number of adults']
#this is how to create or add a new column to the data set, we add the square bracket besides the data variable name and then
#we add the new column title/name between the single quotation
```

```python
data['number of nights per reservation'] = data['number of week nights'] + data['number of weekend nights']
# number of nights per family shows how many days are booked per reservation
```

```python
data['room type'] = data['room type'].replace("Room_Type 1",1)
data['room type'] = data['room type'].replace("Room_Type 2",2)
data['room type'] = data['room type'].replace("Room_Type 3",3)
data['room type'] = data['room type'].replace("Room_Type 4",4)
data['room type'] = data['room type'].replace("Room_Type 5",5)
data['room type'] = data['room type'].replace("Room_Type 6",6)
data['room type'] = data['room type'].replace("Room_Type 7",7)
# transformation from categorical to numerical values
```

```python
data['booking status'] = data['booking status'].replace("Not_Canceled",1)
data['booking status'] = data['booking status'].replace("Canceled",0)
# transformation from categorical to numerical values
```

```python
data['market segment type'] = data['market segment type'].replace("Offline",0)
data['market segment type'] = data['market segment type'].replace("Online",1)
data['market segment type'] = data['market segment type'].replace("Aviation",2)
data['market segment type'] = data['market segment type'].replace("Corporation",3)
data['market segment type'] = data['market segment type'].replace("Complementary",4)
# transformation from categorical to numerical values
```

```python
data['type of meal'] = data['type of meal'].replace("Not Selected",0)
data['type of meal'] = data['type of meal'].replace("Meal Plan 1",1)
data['type of meal'] = data['type of meal'].replace("Meal Plan 2",2)
data['type of meal'] = data['type of meal'].replace("Meal Plan 3",3)
# transformation from categorical to numerical values
```

```python
data[data['date of reservation'].isna()]
```

```python
data['date of reservation'] = pd.to_datetime(data['date of reservation'], errors='coerce').dt.strftime('%m/%d/%Y')
# changes the format of the dates of reservations and making it consistant in the following format "%m/%d/%y"
```

```python
data.head(10)
```

```python
data['date'] = pd.to_datetime(data['date of reservation'], errors='coerce')

data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
#parsing the year and the month from each date of reservation
```
